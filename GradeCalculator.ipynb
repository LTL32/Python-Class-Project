```python
# Grade Calculator Program

# Get user's name Levar
name = input("Enter your name: ")

# Get exam scores Darrin
score1 = float(input("Enter your first exam score: "))
score2 = float(input("Enter your second exam score: "))
score3 = float(input("Enter your third exam score: "))

# Calculate average Levar
average = (score1 + score2 + score3) / 3

# Determine letter grade Darrin
if average >= 90:
    letter = "A"
elif average >= 80:
    letter = "B"
elif average >= 70:
    letter = "C"
elif average >= 60:
    letter = "D"
else:
    letter = "F"

# Display results
print("\n--- Grade Report ---")
print("Student Name:", name)
print("Average Score:", round(average, 2))
print("Letter Grade:", letter)
```
